```
import pandas as pd
```

```
msft = pd.read_csv('MSFT.csv')
msft['name'] = ['MSFT' for i in range(len(msft.index))]
aapl = pd.read_csv('AAPL.csv')
aapl['name'] = ['AAPL' for i in range(len(aapl.index))]
amzn = pd.read_csv('AMZN.csv')
amzn['name'] = ['AMZN' for i in range(len(amzn.index))]
goog = pd.read_csv('GOOG.csv')
goog['name'] = ['GOOG' for i in range(len(goog.index))]

df = msft
df = df.append(aapl)
df = df.append(amzn)
df = df.append(goog)

print(df)
```

```
groups = df.groupby('name') 
# Creates a grouping of the DataFrame by column 'name'
print(groups)
```

```
msft_data = groups.get_group('MSFT')
print(msft_data)
```

```
groups = df.groupby('date') 
# Creates a grouping of the DataFrame by column 'date'
print(groups)
```

```
msft_df = pd.read_csv('MSFT.csv')
msft_df = msft_df[-90:]
amzn_df = pd.read_csv('AMZN.csv')
amzn_df = amzn_df[-90:]
aapl_df = pd.read_csv('AAPL.csv')
aapl_df = aapl_df[-90:]
goog_df = pd.read_csv('GOOG.csv')
goog_df = goog_df[-90:]

msft_close = msft_df['close'].tolist()
amzn_close = amzn_df['close'].tolist()
goog_close = goog_df['close'].tolist()
aapl_close = aapl_df['close'].tolist()
```

```
groups = df.groupby('date')

dates = groups.groups.keys()
avgcloses = []

for date in dates:
  target_data = groups.get_group(date)
  close_data = target_data['close']
  avgclose = close_data.mean()
  avgcloses.append(avgclose)
avgcloses = avgcloses[-90:]
dates = pd.to_datetime(msft_data['date']).tolist()
```

```
df = msft_df.append(aapl_df)
df = df.append(amzn_df)
df = df.append(goog_df)
print(df)

dates = df.groupby('date')
print(dates.get_group('2020-02-05'))
close = []
for date in dates.groups.keys():
  stocks_on_date = dates.get_group(date)
  avg_close_on_date = stocks_on_date['close'].mean()
  close.append(avg_close_on_date)

print(close)
```

```
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()

first = msft_close[0] # Get the value of the first item in the list
msft_close = [((i-first)/first) * 100 for i in msft_close] # Calculate the percentage increase/decrease in price
first = amzn_close[0]
amzn_close = [((i-first)/first) * 100 for i in amzn_close]
first = aapl_close[0]
aapl_close = [((i-first)/first) * 100 for i in aapl_close]
first = goog_close[0]
goog_close = [((i-first)/first) * 100 for i in goog_close]
first = avgcloses[0]
avgcloses = [((i-first)/first) * 100 for i in avgcloses]

stocks = {'MSFT': msft_close,
          'AAPL': aapl_close,
          'AMZN': amzn_close,
          'GOOG': goog_close,
          'AVG': avgcloses
          }

colors = ['#FF0000', '#00FF00', '#0000FF', '#FFFF00', '#000000']
widths = [1, 1, 1, 1, 3]

tech_stocks = figure(title="Tech Stocks", x_axis_label='Date', y_axis_label='Percentage Growth', x_axis_type='datetime')
from bokeh.models import ColumnDataSource

for data, symbol, color, width in zip(stocks.values(), stocks.keys(), colors, widths): # Loop through every stock / color and plot it
  size = len(data) # Get the length of the closing prices list
  symbols = [symbol for i in range(size)] # Create an array of length size, filled with the stock's symbol
  source = ColumnDataSource(data={
      'Date': dates,
      'Close': data,
      'Symbol': symbols
  })
  tech_stocks.line(x='Date', y='Close', source=source, legend_label=symbol, color=color, width=width)

# Display the graph
tech_stocks.legend.location='top_left'
show(tech_stocks)
```
